```python
#This statement - assigns that column to a new variable 
pub = df['Place of Publication']
#This statement assigns a bool label to values of 'london' in the column 
london = pub.str.contains('London')
#This will print the first 5 shows of the data frame
london[:5]
```

```python
# This statement assigns bool label to values of 'oxford' in the column
oxford = pub.str.contains('Oxford')
```

```python
#This statement takes the data frame, replaces anything with london, with singlar london. anything with oxford as oxford
#and will delete anything with a hyphen and replace it with a space 

df['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               pub.str.replace('-', ' ')))
```
